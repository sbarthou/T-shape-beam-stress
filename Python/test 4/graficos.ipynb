```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('data.xlsx')
dimensiones = [6, 8, 10, 12, 14, 16, 20, 22, 25, 28, 32]
```

```python
# filas con el minimo valor de Wpp (tienen el mismo valor)
minimos = df[df['Wpp (ton/m)'] == df['Wpp (ton/m)'].min()]

# menor diferencia entre sigma tension y sigma a
min_dif = minimos[minimos['Dif T-a (ton/m^2)'] == minimos['Dif T-a (ton/m^2)'].min()]
min_dif
```

```python
# mayor diferencia entre sigma tension y sigma a
max_dif = minimos[minimos['Dif T-a (ton/m^2)'] == minimos['Dif T-a (ton/m^2)'].max()]
max_dif
```

```python
# Ancho viga (mm) - Altura viga (mm) | Wpp (ton/m)	
plt.figure(figsize=(10,8))
plt.scatter(df['Ancho viga (mm)'], df['Altura viga (mm)'], s=10, c=df['Wpp (ton/m)'], cmap='viridis_r')
plt.colorbar()
plt.scatter(min_dif['Ancho viga (mm)'], min_dif['Altura viga (mm)'], s=100, marker='*', c='#00EEFF')
plt.scatter(max_dif['Ancho viga (mm)'], max_dif['Altura viga (mm)'], s=100, marker='*', c='red')
plt.title('Ancho viga - Altura viga | Wpp')
plt.xlabel('Ancho viga (mm)')
plt.ylabel('Altura viga (mm)')
plt.show()
```

```python
# Minimos Wpp: mismo valor

minimos = df[df['Wpp (ton/m)'] == df['Wpp (ton/m)'].min()]

plt.figure(figsize=(5,3))
plt.scatter(minimos['Ancho viga (mm)'], minimos['Altura viga (mm)'], s=20)
plt.title('Ancho viga - Altura viga (Wpp minimos)')
plt.xlabel('Ancho viga (mm)')
plt.ylabel('Altura viga (mm)')
plt.show()
```

```python
# Con ancho y alto que dieron el menor valor para Wpp y menor diferencia entre sigma tension y sigma adm. Iterado en dim1 y dim2
df2 = pd.read_excel('data2.xlsx')

plt.figure(figsize=(5,3))
plt.scatter(df2['Ancho alma (mm)'], df2['Altura ala (mm)'], c=df2['Wpp (ton/m)'], cmap='viridis_r')
plt.colorbar()
plt.xticks(dimensiones)
plt.yticks(dimensiones)
plt.title('Ancho alma - Altura ala | Wpp')
plt.xlabel('Ancho viga (mm)')
plt.ylabel('Altura viga (mm)')
plt.show()
```
