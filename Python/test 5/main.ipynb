```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('data.xlsx')
df.head(1)
```

```python
# filas con el minimo valor de Wpp (tienen el mismo valor)
minimos = df[df['Wpp (ton/m)'] == df['Wpp (ton/m)'].min()]

# menor diferencia entre sigma tension y sigma a
min_dif = minimos[minimos['Dif C-a (ton/m^2)'] == minimos['Dif C-a (ton/m^2)'].min()]
min_dif
```

```python
# Ancho viga (mm) - Altura viga (mm) | Wpp (ton/m)	
plt.figure(figsize=(9,7))
plt.scatter(min_dif['Ancho viga (mm)'], min_dif['Altura viga (mm)'], s=100, marker='*', c='red')
plt.scatter(df['Ancho viga (mm)'], df['Altura viga (mm)'], s=10, c=df['Wpp (ton/m)'], cmap='viridis_r')
plt.colorbar()
plt.title('Ancho viga - Altura viga | Wpp')
plt.xlabel('Ancho viga (mm)')
plt.ylabel('Altura viga (mm)')
plt.show()
```
